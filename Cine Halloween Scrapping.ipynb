```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


# Site da Coleta
url = 'https://forbes.com.br/principal/2020/10/os-filmes-de-terror-mais-lucrativos-de-todos-os-tempos/'


# Coletar informações do site
def coletar_infos(url):
    response = requests.get(url)  # Correto: requests.get()
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Coletar todos os <strong> e o texto que vem após
    filmes = []
    valores = []
    for strong_tag in soup.find_all('strong'):
        titulo = strong_tag.text.strip()  # Texto dentro da tag <strong>
        
        # Valores
        valor = strong_tag.next_sibling
        if valor and isinstance(valor, str):
            valor = valor.strip()  # Remover espaços extras
        else:
            valor = ''  # Valor vazio se não houver texto após <strong>
        
        filmes.append(titulo)  # Adiciona o título à lista de filmes
        valores.append(valor)  # Adiciona o valor à lista de valores
        
    return filmes, valores

# Coletar informações
filmes, valores = coletar_infos(url)

# Converter para DataFrame
df = pd.DataFrame({
    'Filmes': filmes, 
    'Valores': valores 
})

# Salvando os dados em uma planilha Excel
df.to_excel("filmes.xlsx", index=False)

print("Dados coletados e salvos em filmes.xlsx")
```
